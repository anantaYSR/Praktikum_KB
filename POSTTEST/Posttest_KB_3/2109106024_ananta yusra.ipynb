```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

*   BARPLOT

```python
df = pd.read_csv("Before_credit_risk_dataset.csv")
x = list(df["person_home_ownership"])
y = list(df["loan_int_rate"])

plt.figure(figsize = (15,5))
plt.bar(x, y, label = "income", color = "red")
plt.title("perbandingan jenis kredit dengan tinggi rate pinjaman")
plt.legend()
plt.show()
```

* dari data di atas kita bisa melihat dari 3000 lebih data yang dimiliki kita dapat melihat berapa persen kemungkinan kita mendapatkan pinjaman berdasarkan jenis kredit nya. dari grafik tersebut maka dapat di simpulkan bahwa dengan jenis kredit Rent(menyewa) kita memiliki kemungkinan mendapatkan pinjaman lebih dari 20%

*   BOXPLOT

```python
sns.boxplot(data = df, y = 'loan_intent', x = 'loan_percent_income')
```

* Dari data di atas dapat di lihat janis kempemilikan Kredit bahwa income yang dimiliki berbeda tergantung jenis kredit yang di ambil, dari data di atas dapat di simpulkan bahwa jenis kepemilikan kredit dengan income terbesar adalah personal dengan income lebih dari 0,8% dari total 100% kepemilakan.

*   SEABORN

```python
sns.set_style("ticks")
plt.figure(figsize = (8,9))
sns.countplot(data = df, y = "loan_grade", hue = "person_home_ownership", palette = "rocket")
```

* Dari grafik di atas kita bisa melihat grade berdasarkan huruf, maka dapat di simpulkan bahwa setiap jenis kredit memiliki grade yang berbeda.

*   PIE CHART

```python
data_pie = df['loan_intent'].value_counts().rename('loan_intent').reset_index(name = 'id_count')
plt.pie(data_pie.id_count, labels = data_pie['index'], startangle = 180, autopct = '%.1f%%', explode = (0.1, 0, 0, 0, 0, 0),
        shadow = True)
plt.title('jenis kepemilikan Kredit')
plt.show()
```

* Jenis kepemilikan Kredit terbesar dari data yang dimiliki adalah dari Education yang di dapat dari seluruh total tiap jenis data kepemilikan pribadi (loan_intent)
