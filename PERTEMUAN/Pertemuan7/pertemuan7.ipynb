# Modul 7 - TENSORFLOW

```python
import tensorflow as tf
```

## Cara membuat Tensor 
## 1. Constant

### A. Constant

```python
const_a = tf.constant([1,2,3,4], shape=[2,2], dtype=tf.float32)
const_a.numpy()
```

### B. Zeros

```python
zeros_b = tf.zeros(shape=[3,2], dtype=tf.int32)
zeros_b.numpy()
```

### C. Fill

```python
fill_c = tf.fill ([4,4], 7)
fill_c.numpy()
```

### D. Random

```python
random_d = tf.random.normal([3,3])
random_d.numpy()
```

```python
random_d = tf.random.uniform([3,3])
random_d.numpy()
```

### E. Convert To Tensor

```python
tensor_e = tf.convert_to_tensor ([[1,3,5,7,9,11],[2,4,6,8,10,12]])
tensor_e.numpy()
```

## 2. Variabel

```python
var1 = tf.Variable([[1,3,5,7],[2,4,6,8]])
#var1 = tf.Variable(tf.ones([3,3]))
var1
```

```python
# mengubah nilai var 1 di atas
var_1_ubah = [[1,2,3,4],[5,6,7,8]]
var1.assign(var_1_ubah)
var1.read_value()
```

# Tensor Slicing & Indexing

## 1. Slicing

```python
tensor_i = tf.random.normal([4,3,3,3])
tensor_i
```

```python
tensor_i[0,...]
```

```python
tensor_i [::1,...]
```

```python
tensor_i [::2,...]
```

## 2. indexing

### A. Gather

```python
indices = [0,1,3]
tf.gather(tensor_i, axis=0, indices=indices)
```

### B. Gather n-Dimension

```python
indices2 = [[0,1,1,0], [1,2,2,0]]
tf.gather_nd(tensor_i, indices=indices2)
```

# Tensor Dimension Modificaton

## 1. Dimension Display

```python
dim_display = tf.constant([[1,2,3,4]], shape=[2,2])
dim_display.shape
```

```python
dim_display.get_shape
```

```python
tf.shape(dim_display)
```

## 2. Dimension Reshaping

```python
dim_reshape = tf.constant([[1,2,3], [4,5,6]])
dim_reshape
```

```python
tf.reshape(dim_reshape, (3,2))
```

## 3. Dimension Ekspansion (perbesaran)

```python
dim_expand = tf.random.normal([100,100,3], seed=1)
dim_expand.shape
```

```python
tf.expand_dims(dim_expand, axis=0).shape
```

```python
tf.expand_dims(dim_expand, axis=1).shape
```

## 4. Dimension Squeezing

```python
dim_squeezing = tf.random.normal([100,1,100,3])
```

```python
dim_squeezing2 = tf.squeeze(dim_squeezing)
dim_squeezing2.shape
```

## 5. Transpose

```python
trp = tf.constant([1,2,3,4,5,6], shape=[2,3])
trp.shape
```

```python
trp2 = tf.transpose(trp)
trp2.shape
```

## 6. Brodcast

```python
brd = tf.constant([1,2,3,4,5,6,7]) #panjang kolom =7
brd
```

```python
brd2 = tf.broadcast_to(brd, shape=[4,7]) #4,7 dimana 7=jumlah kolom dan harus sama dengan panjang kolom di atas
brd2
```

# Aritmatika Tensor

```python
# PENJUMLAHAN

a = tf.constant([[3,5],[6,8]])
b = tf.constant([[1,6],[2,9]])

tf.add(a,b)
# a+b
```

```python
# PERKALIAN

tf.matmul(a,b)
```

```python
# STATISTIKA

argmax = tf.constant([[3,5,1],[6,8,2], [7,5,9]])
argmax
```

```python
max1 = tf.argmax(argmax, axis=0) # Untuk melihat lokasi nilai maksimum per kolom 
max1
```

```python
max2 = tf.argmax(argmax, axis=1) # Untuk melihat lokasi nilai maksimum per baris
max2
```

##  Tensor Concat

### A. Concat

```python
concat1 = tf.random.uniform([4,10,10,3])
concat2 = tf.random.uniform([40,10,10,3])
```

```python
concat1dan2 = tf.concat([concat1,concat2], axis=0)
concat1dan2
```

### B. tf.stack()

```python
stack1 = tf.random.normal([10,10,3])
stack2 = tf.random.normal([10,10,3])
```

```python
stack1dan2 = tf.stack([stack1, stack2], axis=0)
stack1dan2
```

## Tensor Splitting

### A. tf.unstack()

```python
unstack1dan2 = tf.unstack(stack1dan2)
unstack1dan2
```

## Tensor Sorting

### tf.sort()

```python
# TF.SORT

sort1 = tf.random.shuffle(tf.range(10))
sort1
```

```python
sort_asc = tf.sort(sort1, direction="ASCENDING")
sort_asc
```

```python
sort_desc = tf.sort(sort1, direction="DESCENDING")
sort_desc
```

```python
# Menampilkan indexnya
argsort = tf.sort(sort1, direction="ASCENDING")
argsort
```

### tf.nn.top_k

```python
# TF.NN.TOP_K

values, index = tf.nn.top_k(sort1, 5)
values.numpy()
```
